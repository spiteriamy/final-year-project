# **Latin Grammar Question Intent Classifier Interface**

Interface for using the Latin Grammar Question Intent Classifier that allows the user to enter a question and prints the predicted intent.

```python
print(torch.cuda.is_available())
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

## **Load Model**

```python
# load final model and tokenizer
MODEL_PATH = os.path.join('bert_finetuned')

model = AutoModelForSequenceClassification.from_pretrained(MODEL_PATH).to(device)
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
```

## **Prediction Function**

```python
def predict(input: str):
  inputs = tokenizer(input, return_tensors="pt").to(device)
  outputs = model(**inputs)
  prediction = outputs.logits.argmax(dim=1).item()
  label_name = model.config.id2label[prediction]
  return prediction, label_name
```

## **UI**

```python
print("Enter 'q' to exit.\n")

while True:
  # get user input
  text = input("Enter a question: ")

  if text == 'q':
    break

  # get question prediction
  prediction, label = predict(text)

  # print prediction
  print(f"Prediction: {label}")
```
